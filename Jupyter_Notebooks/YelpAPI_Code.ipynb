```python
#VERSION WITH CITY LOOP

from pprint import pprint
import requests
from api_keys import YelpAPI2 as api_key
import pandas as pd

shell = []
results = {'Name': [],'Location':[],'Categories':[], 'Rating': [], 'Price': [], 'Reviews': []}
cities = ["Washington, DC", "Boston, MA", "Newark, NJ", "Baltimore, MD", "Seattle, WA", "Huntsville, AL", "Little Rock, AR", "Charleston, NC", "New Orleans, LA", "Jackson, MS"]
#cities = ["Tallahasse, FL","Jackson, MS"]

for city in cities:
    #print(city)
    url = "https://api.yelp.com/v3/businesses/search"

    headers = {"Authorization": "Bearer {}".format(api_key)}

    params =  {'location': city,
          'term': 'Restaurant',
          'radius': 5000,
          'limit': 50}

    response = requests.get(url, headers=headers, params=params)
    businesses = response.json()['businesses']

    
    for d in businesses:
        #pprint(r)
        results['Name'].append(d['name'])
        results['Location'].append(d['location']['city'])
        results['Categories'].append(d['categories'][0]['title'])
        results['Rating'].append(d['rating'])
        try:
            results['Price'].append(d['price'])
        except:
            results['Price'].append(None)
        results['Reviews'].append(d['review_count'])
        shell.append(results)
    #pprint(response.text)

#pprint(shell)
df = pd.DataFrame(results)

df.to_csv("output_data/Yelp_data.csv", index = False)

df
```
