```python
import pandas as pd
from pathlib import Path
```

```python
# Define the paths to the CSV files for each state
csv_path = Path("Resources/State Affluency.csv")
```

```python
# Read the CSV file
df = pd.read_csv(csv_path)

df
```

```python
#Display only relevant data
filtered_df = df[["Geography", "Estimate (Consumer expenditures per household on Entertainment / Recreation)"]]
```

```python
# Rename the columns 
filtered_df.rename(columns={'Estimate (Consumer expenditures per household on Entertainment / Recreation)': 'Entertainment/Recreation'}, inplace=True)
```

```python
# Sort the DataFrame by the "Entertainment/Recreation" column in descending order
sorted_df = filtered_df.sort_values(by="Entertainment/Recreation", ascending=False)

sorted_df.head()
```
