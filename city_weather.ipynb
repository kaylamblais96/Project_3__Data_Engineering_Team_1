```python
import matplotlib.pyplot as plt
import requests
import pandas as pd
from pprint import pprint
from datetime import datetime, timedelta
```

```python
baltimore_path = "city_data/baltimore.csv"
boston_path = "city_data/boston.csv"
charleston_path = "city_data/charleston.csv"
huntsville_path = "city_data/huntsville.csv"
jackson_path = "city_data/jackson.csv"
little_rock_path = "city_data/little_rock.csv"
new_orleans_path = "city_data/new_orleans.csv"
newark_path = "city_data/newark.csv"
seattle_path = "city_data/seattle.csv"
washington_path = "city_data/washington.csv"
```

```python
baltimore = pd.read_csv(baltimore_path)
boston = pd.read_csv(boston_path)
charleston = pd.read_csv(charleston_path)
huntsville = pd.read_csv(huntsville_path)
jackson = pd.read_csv(jackson_path)
little_rock = pd.read_csv(little_rock_path)
new_orleans = pd.read_csv(new_orleans_path)
newark = pd.read_csv(newark_path)
seattle = pd.read_csv(seattle_path)
washington = pd.read_csv(washington_path)
```

```python
baltimore_df = pd.DataFrame(baltimore)
boston_df = pd.DataFrame(boston)
charleston_df = pd.DataFrame(charleston)
huntsville_df = pd.DataFrame(huntsville)
jackson_df = pd.DataFrame(jackson)
little_rock_df = pd.DataFrame(little_rock)
new_orleans_df = pd.DataFrame(new_orleans)
newark_df = pd.DataFrame(newark)
seattle_df = pd.DataFrame(seattle)
washington_df = pd.DataFrame(washington)
```

```python

#An example of one of the dataframes
washington_df.head()
```

```python
#Removed some of the unnecessary columns

baltimore_df = baltimore_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
boston_df = boston_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
charleston_df = charleston_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
huntsville_df = huntsville_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
jackson_df = jackson_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
little_rock_df = little_rock_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
new_orleans_df = new_orleans_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
newark_df = newark_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
seattle_df = seattle_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
washington_df = washington_df.loc[:, ['name', 'datetime', 'tempmin', 'tempmax', 'sunrise', 'sunset', 'icon']]
```

```python
washington_df
```

```python
#Merged all of the DataFrames into one

merged_df = pd.concat([baltimore_df, boston_df, charleston_df, huntsville_df, jackson_df, little_rock_df, 
                       new_orleans_df, newark_df, seattle_df, washington_df])

merged_df.reset_index(drop=True, inplace=True)

merged_df
```

```python
#Added a new column to the dataframe with the 'average' daily temp.

merged_df['daily_avg_temp'] = (merged_df['tempmin'] + merged_df['tempmax']) / 2
merged_df
```

```python
#Added a month column to help if we graph or anything by month

merged_df['datetime'] = pd.to_datetime(merged_df['datetime'])
merged_df['month'] = merged_df['datetime'].dt.month
merged_df
```
